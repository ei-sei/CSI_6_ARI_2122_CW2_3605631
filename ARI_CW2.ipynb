##handling categorical variables

```python
df.head()
```

```python
from sklearn import preprocessing
#create an instance of the label encoder and name it as 'encoder'
encoder = preprocessing.LabelEncoder()
```

```python
# encoder.fit_transform(df["CustomerType"])
# encoder.fit_transform([])
```

```python
encoder.fit(df['satisfaction'])
df.satisfaction = encoder.transform(df.satisfaction)
```

```python
encoder.fit(df['Gender'])
df.Gender = encoder.transform(df.Gender)
```

```python
encoder.fit(df['CustomerType'])
df.CustomerType = encoder.transform(df.CustomerType)
```

```python
encoder.fit(df['TypeOfTravel'])
df.TypeOfTravel = encoder.transform(df.TypeOfTravel)
```

```python
encoder.fit(df['Class'])
df.Class = encoder.transform(df.Class)
```

```python
df.head()
```

##train split and scaling data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier

```

```python
# import sklearn
```

```python
X = df.drop('satisfaction',axis=1)
y = df['satisfaction']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```
